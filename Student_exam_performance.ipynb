```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.linear_model import LinearRegression

data = pd.read_csv('student_exam_performance.csv')

# 2. Prepare input (X) and output (y)
X = data[['Study_Hours']]        
y = data['Final_Exam_Score']       


model = LinearRegression()
model.fit(X, y)

# 4. Save the trained model
joblib.dump(model, 'student_score_model_real.pkl')
print("✅ Model saved as 'student_score_model_real.pkl'")

# 5. Generate and save the performance plot
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='black', label='Actual Scores')
plt.plot(X, model.predict(X), color='blue', linewidth=2, label='Predicted Line')
plt.xlabel('Study Hours')
plt.ylabel('Final Exam Score')
plt.title('Student Exam Performance Prediction (Real Data)')
plt.legend()
plt.grid(True)
plt.savefig('performance_chart_real.png')
plt.show()
```
